```python
from netCDF4 import Dataset                     # For reading data
import matplotlib.pyplot as plt                 # For plotting
# Place figures within document
%pylab inline
pylab.rcParams['figure.figsize'] = (18.0, 10.0) # Large figures
```

```python
# Load data
visc = Dataset('visc.nc')
time = visc.variables['Time'][:]
h = visc.variables['KPP_OBLdepth'][:,0,0]
Kt = visc.variables['KPP_Kheat'][:,:,0,0]
Ku = visc.variables['KPP_Kv'][:,:,0,0]
#for v in visc.variables: print v
prog  = Dataset('prog.nc')
#for v in prog.variables: print v
zi = prog.variables['zi'][:]
zl = prog.variables['zl'][:]
u = prog.variables['u'][:,:,0,0]
v = prog.variables['v'][:,:,0,0]
temp = prog.variables['temp'][:,:,0,0]
```

```python
plt.subplot(2,1,1); plt.plot(time, -h,'-'); plt.ylim((-20,0))
plt.ylabel('z (m)'); plt.title('KPP boundary layer depth - test case WSwPSBF')
plt.subplot(2,1,2); plt.plot(time, temp[:,0],'-'); plt.ylim((14.8,17.0))
plt.xlabel('Time (days)'); plt.ylabel(r'$\theta$ (${\degree}C$)')
plt.title('SST');
```

```python
plt.subplot(2,1,1); plt.contourf(time, -zi, Kt.T, levels=numpy.arange(0,.021,.001)); plt.colorbar()
plt.ylim((-50,0)); plt.ylabel('z (m)'); plt.title(r'$\kappa_\theta (m^2/s)$')
plt.subplot(2,1,2); plt.contourf(time, -zi, Ku.T, levels=numpy.arange(0,.021,.001)); plt.colorbar()
plt.ylim((-50,0)); plt.ylabel('z (m)'); plt.title(r'$\kappa_u (m^2/s)$'); plt.xlabel('Time (days)');
```
